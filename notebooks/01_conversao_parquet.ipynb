# 01 – Conversão CSV para Parquet  

Converte todos os CSVs da pasta `raw/` para o formato Parquet e os salva na pasta `bronze/`, junto com os arquivos JSON de metadados.


```python
import pandas as pd
import shutil
from pathlib import Path

RAW_DIR = Path("../raw")
BRONZE_DIR = Path("../bronze")
BRONZE_DIR.mkdir(exist_ok=True)
```

```python
csv_files = list(RAW_DIR.glob("*.csv"))
print(f"CSVs encontrados: {[f.name for f in csv_files]}")
```

```python
for csv_path in csv_files:
    print(f"\n--- {csv_path.name} ---")
    
    # Ler CSV
    df = pd.read_csv(csv_path)
    print(f"  Linhas: {len(df):,}, Colunas: {len(df.columns)}")
    
    # Salvar como Parquet
    parquet_name = csv_path.stem + ".parquet"
    parquet_path = BRONZE_DIR / parquet_name
    df.to_parquet(parquet_path, index=False)
    print(f"  → Parquet salvo: {parquet_path.name}")
    
    # Copiar metadados JSON associado
    json_path = csv_path.with_suffix(".json")
    if json_path.exists():
        shutil.copy(json_path, BRONZE_DIR / json_path.name)
        print(f"  → Metadados copiados: {json_path.name}")
    else:
        print(f"  → ⚠️ Metadados não encontrado para {csv_path.name}")

print("\n✅ Conversão concluída.")
```
